## 2. Data Cleaning and Missing Values (14pts)

* Check for missing values across all columns.

* Which columns contain missing data, and how many missing entries are there?

* What strategies could you use to handle them (drop, fill, or ignore)? Apply one reasonable approach and explain your choice.

```python
# Keep column names exactly as in file (just trim spaces)
df.columns = df.columns.str.strip()

print(f"Shape BEFORE cleaning: {df.shape[0]} rows × {df.shape[1]} columns\n")

# --- 1. Check for missing values across all columns ---
missing_values = df.isna().sum()
missing_cols = missing_values[missing_values > 0]

print("=== Missing Values Summary ===")
if missing_cols.empty:
    print("No missing values detected.")
else:
    print(missing_cols)
    print("\n(% of rows missing per column)")
    print(((missing_cols / len(df)) * 100).round(2))

# --- 2. Apply reasonable strategy (if missing exist) ---
# Strategy: numeric → median, categorical → mode
df_clean = df.copy()

num_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df_clean.select_dtypes(exclude=[np.number]).columns.tolist()

for c in num_cols:
    if df_clean[c].isna().any():
        med = df_clean[c].median()
        df_clean[c] = df_clean[c].fillna(med)
        print(f"Filled numeric column '{c}' with median value: {med:.2f}")

for c in cat_cols:
    if df_clean[c].isna().any():
        mode_val = df_clean[c].mode(dropna=True)
        fill_val = mode_val.iloc[0] if not mode_val.empty else "Unknown"
        df_clean[c] = df_clean[c].fillna(fill_val)
        print(f"Filled categorical column '{c}' with mode: {fill_val}")

# --- 3. Verify imputation worked ---
remaining_na = int(df_clean.isna().sum().sum())
print(f"\nRemaining missing cells after imputation: {remaining_na}")

# --- 4. Validate column integrity ---
print("\nColumn count before vs after cleaning:")
print("Before:", len(df.columns))
print("After :", len(df_clean.columns))

# Check if any column accidentally dropped or renamed
if set(df.columns) != set(df_clean.columns):
    print("\n⚠️ Columns missing after cleaning:")
    print(set(df.columns) - set(df_clean.columns))
else:
    print("\n All columns preserved correctly.")

print(f"\n Final shape AFTER cleaning: {df_clean.shape[0]} rows × {df_clean.shape[1]} columns")
```

After reloading and trimming column names, the dataset contains 2,240 rows and 29 columns.
A check for missing values shows no missing data across any variable.
If missing values had existed, numeric fields (e.g., Income, Recency) would be imputed using the median, while categorical fields (e.g., Education, Marital_Status) would use the mode to maintain realistic distributions.
All 29 columns are preserved and verified, leaving the dataset fully clean and ready for analysis.
