## 2. Data Cleaning and Missing Values (14pts)

* Check for missing values across all columns.

* Which columns contain missing data, and how many missing entries are there?

* What strategies could you use to handle them (drop, fill, or ignore)? Apply one reasonable approach and explain your choice.

```python
# Keep column names exactly as in file (just trim spaces)
df.columns = df.columns.str.strip()

print(f"Shape BEFORE cleaning: {df.shape[0]} rows × {df.shape[1]} columns\n")

# --- 1. Check for missing values across all columns ---
missing_values = df.isna().sum()
missing_cols = missing_values[missing_values > 0]

print("=== Missing Values Summary ===")
if missing_cols.empty:
    print("No missing values detected.")
else:
    print(missing_cols)
    print("\n(% of rows missing per column)")
    print(((missing_cols / len(df)) * 100).round(2))

# reasonable strategy (if missing exists)
# Strategy: numeric → median, categorical → mode
df_clean = df.copy()

num_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df_clean.select_dtypes(exclude=[np.number]).columns.tolist()

for c in num_cols:
    if df_clean[c].isna().any():
        med = df_clean[c].median()
        df_clean[c] = df_clean[c].fillna(med)
        print(f"Filled numeric column '{c}' with median value: {med:.2f}")

for c in cat_cols:
    if df_clean[c].isna().any():
        mode_val = df_clean[c].mode(dropna=True)
        fill_val = mode_val.iloc[0] if not mode_val.empty else "Unknown"
        df_clean[c] = df_clean[c].fillna(fill_val)
        print(f"Filled categorical column '{c}' with mode: {fill_val}")

# --- 3. Verify imputation worked ---
remaining_na = int(df_clean.isna().sum().sum())
print(f"\nRemaining missing cells after imputation: {remaining_na}")

# --- 4. Validate column integrity ---
print("\nColumn count before vs after cleaning:")
print("Before:", len(df.columns))
print("After :", len(df_clean.columns))

# Check if any column accidentally dropped or renamed
if set(df.columns) != set(df_clean.columns):
    print("\nColumns missing after cleaning:")
    print(set(df.columns) - set(df_clean.columns))
else:
    print("\n All columns preserved correctly.")

print(f"\n Final shape AFTER cleaning: {df_clean.shape[0]} rows × {df_clean.shape[1]} columns")
```

After reloading and trimming column names, the dataset contains 2,240 rows and 29 columns.
A check for missing values shows no missing data across any variable.
If missing values had existed, numeric fields (e.g., Income, Recency) would be imputed using the median, while categorical fields (e.g., Education, Marital_Status) would use the mode to maintain realistic distributions.
All 29 columns are preserved and verified, leaving the dataset fully clean and ready for analysis.

## 4. Add a new column (15pts)

* Add a new column **Customer_Days** to store the total number of days each customer has been with the company. You may follow the steps below:
  * use `pd.to_datetime(df["Dt_Customer"])` to convert the string variable to a *date-time* variable

  * use `today = datetime.today()` to find the date of today

  * use `(today - df["Dt_Customer"]).dt.days` to find the difference between today and the onbard date.

```python
from datetime import datetime
import pandas as pd

# Work on the dataset from previous step
df_days = df_constant.copy()

# Convert 'Dt_Customer' to datetime
df_days['Dt_Customer'] = pd.to_datetime(df_days['Dt_Customer'], errors='coerce')

# Get today's date
today = datetime.today()

# Calculate the number of days each customer has been with the company
df_days['Customer_Days'] = (today - df_days['Dt_Customer']).dt.days

# Verify the result
print("Added 'Customer_Days' column successfully!\n")
print(df_days[['Dt_Customer', 'Customer_Days']].head())

# check for negative or missing values
print("\nMissing Customer_Days:", df_days['Customer_Days'].isna().sum())
print("Minimum days:", df_days['Customer_Days'].min())
print("Maximum days:", df_days['Customer_Days'].max())

# Optional: check a summary
df_days['Customer_Days'].describe().apply(lambda x: format(x, '.0f'))
```

## 5. Categorical Features (15 pts)

* For `Education` and `Marital_Status`,

  * find how many customers fall into each category
  * create pie charts to show the percentage of each category; make sure to show the percentage numbers on the chart with 2 decimal places

* What do you findings suggest about the company's target demographic?

```python
import matplotlib.pyplot as plt
import pandas as pd

# Work on the dataset with Customer_Days included
df_cat = df_days.copy()

# Education - count and percentage
edu_counts = df_cat['Education'].value_counts()
edu_pct = (edu_counts / edu_counts.sum()) * 100

print("=== Customer Counts by Education ===")
print(edu_counts)
print("\n=== Percentage by Education (2 decimal places) ===")
print(edu_pct.round(2))

# Pie chart for Education
plt.figure(figsize=(7,7))
plt.pie(
    edu_pct,
    labels=edu_pct.index,
    autopct='%1.2f%%',
    startangle=140,
    shadow=True
)
plt.title('Distribution of Customers by Education Level')
plt.axis('equal')
plt.show()

# Marital Status - count and percentage
ms_counts = df_cat['Marital_Status'].value_counts()
ms_pct = (ms_counts / ms_counts.sum()) * 100

print("\n=== Customer Counts by Marital Status ===")
print(ms_counts)
print("\n=== Percentage by Marital Status (2 decimal places) ===")
print(ms_pct.round(2))

# Pie chart for Marital Status
plt.figure(figsize=(7,7))
plt.pie(
    ms_pct,
    labels=ms_pct.index,
    autopct='%1.2f%%',
    startangle=140,
    shadow=True
)
plt.title('Distribution of Customers by Marital Status')
plt.axis('equal')
plt.show()
```

The findings suggest that most customers are well-educated, with over half holding a Graduation or PhD degree, indicating that Nata Supermarkets’ target demographic skews toward higher-education, professional households.
In terms of marital status, the majority are Married or Living Together, suggesting that Nata primarily serves family-oriented customers.
This insight can help design promotions emphasizing household and premium lifestyle products.

## 7. Spending by Products (15pts)

* Examine the spending amount columns `MntWines`,	`MntFruits`,	`MntMeatProducts`,	`MntFishProducts`,	`MntSweetProducts`,	`MntGoldProds`, and find the product with the highest total spending amount by all customers.



```python
import matplotlib.pyplot as plt
import pandas as pd

# Work on the latest dataset
df_spend = df_days.copy()

# Define the spending columns
spend_cols = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]

# Calculate total spending per category (sum across all customers)
total_spend = df_spend[spend_cols].sum().sort_values(ascending=False)

# Display results
print("=== Total Spending by Product Category ===")
print(total_spend)

# Find product with highest total spending
top_product = total_spend.idxmax()
top_value = total_spend.max()
print(f"\n The highest total spending is on '{top_product}' with a total of {top_value:.2f} units.\n")

# Visualize spending by category
plt.figure(figsize=(8,6))
plt.bar(total_spend.index, total_spend.values, color='skyblue')
plt.title('Total Spending by Product Category')
plt.ylabel('Total Spend (monetary units)')
plt.xlabel('Product Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
